```python
import numpy as np
import pandas as pd
import matplotlib. pyplot as plt
import csv
import seaborn as sns
```

Uploading the dataset (.csv files)

```python
employeeNumber = 'datasets/employment.csv'
```

Reading the datasets(.csv files)

```python
employeeNumber = pd.read_csv(employeeNumber)
```

Renaming Column Names

```python
employeeNumber.rename(columns={'year':'Year','capital':'NCR','administrative':'CAR','region':'Region I','region_2': 'Region II', 'region_3':'Region III', 'region_4':'Region IV-A', 'region_5':'Region IV-B', 'region_6':'Region V'
                                       , 'region_7':'Region VI','region_8': 'Region VII','region_9': 'Region VIII', 'region_10':'Region IX','region_11': 'Region X', 'region_12':'Region XI', 'region_13':'Region XII'
                                       ,'region_14': 'Region XIII', 'armm':'ARMM'}, inplace=True)
employeeNumber
```

Removing first column of the table

```python
employeeNumber.drop('philippines', axis=1, inplace=True)

employeeNumber
```

```python
plt.figure(figsize=(12, 8))

plt.bar(employeeNumber['Year'], employeeNumber['NCR'], color=['#fbf8cc', '#ffcfd2', '#cfbaf0', '#a3c4f3', '#90dbf4', '#8eecf5', '#98f5e1', '#b9fbc0'])

plt.title("Employment Status of the Philippines in year 2012 to 2019")
plt.xlabel('Year', fontsize=14)
plt.ylabel('National Capital Region', fontsize=14)
plt.grid(True)

plt.show()
```

```python
# Use describe() to get descriptive statistics for all columns
description = employeeNumber.describe()
plt.figure(figsize=(12, 8))
# Plotting the mean values
description.loc['mean'].plot(kind='bar')

# Adding labels and title
plt.xlabel('Columns')
plt.ylabel('Mean')
plt.title('Mean Values of Columns')

# Display the plot
plt.show()
```

```python
description = employeeNumber.describe()

# Extract the statistics you want to visualize
statistics = description.loc[['count', 'mean', 'std', 'min', '25%', '50%']]

# Transpose the DataFrame for easier plotting
statistics = statistics.T

# Adjust figure size
plt.figure(figsize=(12, 8))

# Plotting the line graph
statistics.plot(kind='line', marker='o')

# Adding labels and title
plt.xlabel('Columns')
plt.ylabel('Value')
plt.title('Descriptive Statistics')

# Adjust font size
plt.xticks(range(len(statistics.index)), statistics.index, fontsize=8)
plt.xticks(rotation=90)
# Display the plot
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
